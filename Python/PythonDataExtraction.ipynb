```python
import yfinance as yf
import pandas as pd
```

```python
# Create a ticker object for Blue Apron
blue_apron = yf.Ticker("APRN")
```

```python
# Fetch the income statement data
income_statement = blue_apron.financials

# Show the data to ensure it's correctly fetched
print(income_statement)

# Optionally, save the data to a CSV file for future use
income_statement.to_csv("blue_apron_income_statement.csv")
```

```python
beyond_meat = yf.Ticker("BYND")
```

```python
# Fetch the income statement data
income_statement = beyond_meat.financials

# Show the data to ensure it's correctly fetched
print(income_statement)

# Optionally, save the data to a CSV file for future use
income_statement.to_csv("beyond_meat_income_statement.csv")
```

```python
# Fetch the balance sheet data
balance_sheet = beyond_meat.balance_sheet

# Show the data to ensure it's correctly fetched
print(balance_sheet)

# Optionally, save the data to a CSV file for further use
balance_sheet.to_csv("beyond_meat_balance_sheet.csv")
```

```python
# Fetch the cash flow statement data
cash_flow = beyond_meat.cashflow

# Show the data to ensure it's correctly fetched
print(cash_flow)

# Optionally, save the data to a CSV file for further use
cash_flow.to_csv("beyond_meat_cash_flow.csv")
```

```python
# Get some basic financial ratios and metrics
market_cap = beyond_meat.info.get('marketCap', 'N/A')
pe_ratio = beyond_meat.info.get('trailingPE', 'N/A')
dividend_yield = beyond_meat.info.get('dividendYield', 'N/A')
return_on_equity = beyond_meat.info.get('returnOnEquity', 'N/A')

# Display the results
print(f"Market Cap: {market_cap}")
print(f"P/E Ratio: {pe_ratio}")
print(f"Dividend Yield: {dividend_yield}")
print(f"Return on Equity: {return_on_equity}")
```

```python
# Save financial data to CSV
income_statement.to_csv("beyond_meat_income_statement.csv")
balance_sheet.to_csv("beyond_meat_balance_sheet.csv")
cash_flow.to_csv("beyond_meat_cash_flow.csv")
```
